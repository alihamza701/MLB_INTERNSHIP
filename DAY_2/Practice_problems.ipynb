# **LIST**

```python
#Find the largest number in a list.

mylist = [10,1,60,20,33,19,45,10]
checker=0
for x in mylist:
  if x > checker:
    checker = x

print("Largest number in the list: " , checker)
```

```python
#Find the second largest number.

largest = 0
sec_largest = 0
for x in mylist:
  if x > largest:
    largest = x

for x in mylist:
  if x > sec_largest and x < largest:
    sec_largest = x

print("Second Largest number in the list: " , sec_largest)
```

```python
#Remove duplicate values from a list.

y=[]
for x in mylist:
  if x not in y:
    y.append(x)

mylist = y
print(mylist)
```

```python
#Reverse a list without using built-in reverse().

new_list=[]
i=1
for x in mylist:
  new_list.append(mylist[-i])
  i+=1

print("List in true order: " , mylist)

mylist = new_list

print("List in reverse order :" , mylist)
```

```python
#Find common elements between two lists.

newlist = [19,20,3,88,47,6,23]
common = []

for x in mylist:
  for y in newlist:
    if x==y:
      common.append(x)

print(common)
```

# **TUPLE**

```python
#Count occurrences of an element.

counter=0
print("Values  :   Counter\n")
mytuple = (10, 14, 60, 20, 33, 19, 45, 10)
for x in mytuple:
  for y in mytuple:
    if x==y:
      counter +=1
  print(f"\'{x}\'  :  {counter}")
  counter = 0
```

```python
#Convert a tuple into a list and vice versa.

print("Tuple : " , mytuple)

newlist = list(mytuple)

print("Tuple in a List  format: " , newlist)

mytuple = tuple(newlist)

print("List an a tuple format : " , mytuple)

```

# **SETS**

```python
#Find unique values from a list.


newlist = [19,20,3,88,47,6,23,3,88]
print("List : " , newlist)
for x in newlist:
  for y in newlist:
    if x==y:
      newlist.remove(x)

print("Unique values from the list : " , newlist)
```

```python
#Perform union and intersection operations.

myset = {19, 20, 3, 88, 47, 6, 23}
newset = {10,1,60,20,33,19,45}

print("Set 1 : " ,myset)
print("\nSet 2 : " , newset)

union_set = myset.union(newset)

print("\nUnion of two sets : " , myset)

intersection_set = myset.intersection(newset)

print("\nIntersection of two sets : " , intersection_set)


#Another method

union_set = myset | newset

myset &= newset
```

# **DICTIONARY**

```python
#Create a student record dictionary.

new_dictionary = {}

print("Press Q to quit")

while True:\

  name = input ("Name : ")

  if name == "Q":
    break

  marks = int(input("Marks : "))

  if marks == "Q":
    break

  new_dictionary.update({name : marks})

print(new_dictionary)


```

```python
#Calculate average marks of students.

total = 0
for x in new_dictionary.values():
  total += x
  Avg = total/len(new_dictionary)

print("Marks Average : " , Avg)
```

```python
#Count frequency of words in a sentence.

sentence = "I am Ali Hamza. I am from Lahore , Studying AI at UMT."

words = sentence.split()

dictionary = {}

for x in words:

  if x in dictionary:
    dictionary[x] += 1
  else:
    dictionary[x] = 1

print("*Words Frequencies*")

for x,y in dictionary.items():
    print(f"{x} : {y}")
```
