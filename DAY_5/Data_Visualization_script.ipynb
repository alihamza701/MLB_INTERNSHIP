# **Data Visualization**


**Create the following charts:**

**Bar chart showing the average marks of each student.**

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,4))
plt.bar(df["Name"] , df["Average_Score"])
plt.xlabel("Name")
plt.ylabel("Average_Score")
plt.title("Bar Chart")
plt.tight_layout()
plt.xticks(rotation=90 , fontsize = 16 , color = "red")
plt.yticks(fontsize = 16 , color = "green")
plt.show()
```

**Histogram showing the distribution of Average Scores.**


```python
plt.hist(df["Average_Score"] , bins=10 , color = 'green')
plt.xlabel("Average_Score")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

**Scatter plot comparing Python and Machine Learning marks.**

```python
plt.scatter(df["Python"], df["Machine_Learning"], color = "violet")
plt.xlabel("Python")
plt.ylabel("Machine Learning")
plt.title("Scatter Plot")
plt.show()
```

**Pie chart showing the distribution of students by Performance category.**

```python
performance_counts = df["Performance"].value_counts()
plt.pie(performance_counts, labels=performance_counts.index, colors=plt.cm.Paired.colors)
plt.title("Distribution of Students by Performance Category")
plt.show()
```

**Box plot to visualize the spread of marks in all subjects.**

```python

df.iloc[:,6:12].plot(kind="box", figsize=(8,5))

plt.title("Box Plot")
plt.ylabel("Marks")
plt.show()
```
