#**Student Performance Dashboard**

**Using the dataset, create a simple analysis notebook or Python script that answers the following questions:**

**How many students are in the dataset?**

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("/content/student_performance.xlsx")
```

```python
print(df.columns)
tot_stu = df["Student_ID"].count()
print("Total Students in the class : " , tot_stu)
```

**What is the average score for each subject?**

```python
df["Avg_Score"] = df.iloc[: , 6:12].mean(axis=1)
```

```python

print(df[["Name" , "Avg_Score"]])
```

**Who are the Top 5 performing students?**

```python
df["Avg_Score"].sort_values(ascending=False).head()
```

**Which students need improvement?**

```python
imp = df[df["Avg_Score"]<70]
print(imp)
```

**Which subject has the highest class average?**

```python
high_avg = df.iloc[: , 6:12].mean()
print(high_avg)
print("\n----------------------------------------\n")
print("Subject with the highest average : " , high_avg.idxmax())
```

**Visualize your findings using appropriate charts.**

```python
#Visualizing the avg marks
plt.pie(high_avg , labels=high_avg.index , autopct="%1.1f%%")
plt.title("Average Marks")
plt.show()
```

```python
top_5 = df.sort_values(by = "Avg_Score" , ascending=False).head(5)

plt.bar(top_5["Name"] , top_5["Avg_Score"])
plt.xlabel("Student Names")
plt.ylabel("Average Score")
plt.title("Top 5 Students")
plt.tight_layout()
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(imp["Name"] , imp["Avg_Score"])
plt.xlabel("Student Names")
plt.ylabel("Average Score")
plt.title("Students Needing Improvement (Average Score < 70)")
plt.xticks(rotation=90, fontsize=8)
plt.tight_layout()
plt.show()
```

```python
plt.plot(high_avg)
plt.xlabel("Subjects")
plt.ylabel("Average Score")
plt.title("Subject of highest Average Scores")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
