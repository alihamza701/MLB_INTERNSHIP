```python
import numpy as np

X = np.random.rand(2000,10)
y = np.random.randint(2,size=(2000,1))
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(12,input_shape=(10,),activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

model.compile(
    optimizer = 'adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
model.summary()
```

```python
model.fit(X,y,epochs=10,batch_size=32)
```

```python
new_data = np.random.rand(1,10)

predictions = model.predict(new_data)
print(f"Model Prediction : {predictions[0][0]}")
```

Using sigmoid in Hidden Layers

```python
model2 = Sequential()
model2.add(Dense(12,input_shape=(10,),activation='sigmoid'))
model2.add(Dense(10,activation='sigmoid'))
model2.add(Dense(1,activation='sigmoid'))

model2.compile(
    optimizer = 'adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
model2.summary()
```

```python
model2.fit(X,y,epochs=10,batch_size=32)
```

```python
predictions = model2.predict(new_data)
print(f"Model Prediction : {predictions[0][0]}")
```
