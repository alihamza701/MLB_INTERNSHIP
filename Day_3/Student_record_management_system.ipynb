```python
import json
import os

dictionary = {}

def load():
  if not os.path.exists("Student_rec.json"):
    with open("Student_rec.json","w") as f:
      json.dump({},f)
    return {}
  else:
    with open("Student_rec.json","r") as f:
      return json.load(f)

def save(data):
  with open("Student_rec.json","w") as f:
    json.dump(data,f , indent=2)

  print("Query executed successfully")


def add_func():

  name = input ("Name :")
  age = int(input("Age: "))
  roll_no = input("Roll No: ")
  course  = input("Course : ")
  dictionary[roll_no] = [name,age,course]
  save(dictionary)

def searchh():
  search = input("Enter roll_no :")
  data = load()
  if search in data.keys():
    print(data[search])
  else:
    print("No record found")

def update():

    search = input("Enter roll_no :")
    data = load()
    if search in data.keys():
        name = input ("Name :")
        age = int(input("Age: "))
        course  = input("Course : ")
        dictionary[search] = [name,age,course]
        data.update(dictionary)
        save(data)

    else:
        print("No record found")

def delete():
    search = input("Enter roll_no :")
    data = load()
    if search in data:
      del data[search]
      save(data)
    else:
        print("No record found")

```

```python


while True:

  choice  = int(input("1:Add Student\n2:View All Students\n3:Search Student\n4:Update Student Information\n5:Delete Student\n6. Quit\n"))

  if choice == 1:
    add_func()

  elif choice == 2:

    print(f"Total Students: {len(dictionary.keys())}\n")
    for x,y in dictionary.items():
      print("------------------------")
      print(f"{x} : {y}")

  elif choice == 3:

    searchh()

  elif choice == 4:

    update()

  elif choice == 5:

    delete()

  elif choice == 6:
    break
  else:
    print("Invalid Input")
```
