# **FILE HANDLING**

**Create a text file and write data into it.**

```python

data = "Ali Hamza is a student.\nAt UMT\nIn 8th semester."

with open("newfile.txt" , "w") as file:
  file.write(data)
```

**Readand display file contents.**

```python
with open("newfile.txt" , "r") as file:
  reading_file = file.read()

print("Contents of file :" , reading_file)
```

**Append new data to an existing file.**

```python
with open("newfile.txt" , "a") as file:
  file.write("\nOh! i forgot to append some content")
```


**Count the number of lines in a file.**


```python


count = 0

with open("newfile.txt" , "r") as obj:
  lines = obj.readlines()
  count = len(lines)

print("\nNumber of lines in file :" , count)
```

# **JSON**

**Store student information in a JSON file.**

```python
name = input("Name : ")
age = int(input("Age : "))
course  = input("Course : ")

student_rec  = {
    "Name" : name ,
    "Age" : age ,
    "Course" : course
}

import  json
with open("json_file.json" , "w") as file_obj:

    json.dump(student_rec , file_obj)

print("\n\nData added to json file successfully")
```

**Read data from a JSON file.**

```python



with open("json_file.json" , "r") as file_obj:

  json_data = json.load(file_obj)

print("Reading Data from json file : " , json_data)
```

**Update an existing student's information.**

```python

with open("json_file.json" , "r") as file_obj:

  data = json.load(file_obj)

print(data)

data["Name"] = "Hussain"

with open("json_file.json" , "w") as file_obj:

    json.dump(data , file_obj)

print("\n\nData updated to json file successfully")


with open("json_file.json" , "r") as file_obj:

  data = json.load(file_obj)

print("\n\nUpdated json : " , data)
```

**Add a new student to the JSON file.**

```python
student_rec = {
    "Student1" :
     {"Name" : "Zaigham ali" ,
    "Age" : 20 ,
    "Course" : "AI"}
}

with open("json_file.json" , "w") as file:

  json.dump(student_rec , file)


new_student = {
    "Student2" : {
        "Name" : "Zaigham" ,
        "Age" : 26 ,
        "Course" : "Computer"
    }
}

with open("json_file.json" , "r") as file:
    data = json.load(file)

data.update(new_student)

with open("json_file.json" , "w") as file:
  json.dump(data , file)

with open("json_file.json" , "r") as file:
    data = json.load(file)

print("\n\nUpdated json : " , data)
```
