**Load the dataset using Pandas.**

```python
import pandas as pd

df = pd.read_csv('/content/student_performance.csv')
```

**Display basic information about the dataset.**

```python
print(df.shape)
print("\n----------------------------------------\n")
print(df.head())
print("\n----------------------------------------\n")
print(df.tail())
print("\n----------------------------------------\n")
print(df.describe())
print("\n----------------------------------------\n")
print(df.info())
```

**Calculate average marks for each subject.**

```python
Math = df['Math'].mean()
eng = df['English'].mean()
comp_Sci = df['ComputerScience'].mean()
Sci = df['Science'].mean()

print("Average for Math : " , Math)
print("Average for English : " , eng)

print("Average for Computer Science : " , comp_Sci)
print("Average for Science : " , Sci )
```

**Identify the top 5 performing students.**

```python

df['Total'] = df.iloc[:,2:6].sum(axis=1)
top_5 = df.sort_values('Total' , ascending=False)[:5]
print(top_5[['Name' , 'Total']])
```

**Find students scoring below the average.**

```python

df['Total'] = df.iloc[:,2:6].mean(axis=1)
below_avg = df[df['Total'] < df['Total'].mean()]
print(below_avg[['Name' , 'Total']])
```

**Display the total number of students.**

```python
print("Total Number of Students : " , df['Name'].count())
```
